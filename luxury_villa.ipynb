### Handling Null entries :

### Question 1: Best Price per Sqft Value in City?

```python
city_value=(villa_dataset.groupby("City")["Price_Per_Sqft"].mean().sort_values())

## VISUALIZATION
setup_chart=("Best Luxury Villa Value by City")

city_value.plot(kind="barh",
               edgecolor="white")

plt.xlabel("Average Price Per Sqft")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```python
print(city_value)
```

### Insights:
      Pune = emerging luxury investment hotspot
      Mumbai = ultra-premium saturated market
      Bangalore = balanced appreciation + demand market

### Question 2: Which Amenity Adds most Value?

```python
amenities = [
    "Swimming_Pool",
    "Gym",
    "Garden"
]

results = []

for amenity in amenities:

    yes_price = villa_dataset[
        villa_dataset[amenity].astype(str).str.lower().isin(
            ["yes", "1", "true", "available"]
        )
    ]["Price_Crore"].mean()

    no_price = villa_dataset[
        villa_dataset[amenity].astype(str).str.lower().isin(
            ["no", "0", "false", "not available"]
        )
    ]["Price_Crore"].mean()

    premium = yes_price - no_price

    results.append([amenity, premium])

premium_df = pd.DataFrame(
    results,
    columns=["Amenity", "Premium"]
)


### VISUALIZATION


plt.figure(figsize=(10,5))

sns.barplot(
    data=premium_df,
    x="Amenity",
    y="Premium"
)

plt.title(
    "Luxury Amenity Premium Analysis",
    fontsize=18,
    fontweight="bold"
)

plt.xlabel("Amenity")
plt.ylabel("Average Price Premium")

plt.tight_layout()

plt.show()
```

### Insights:
        1. Villas with premium lifestyle amenities command higher market value.
        2. Swimming pool contributes strongest luxury perception.

### Question 4: Ready to Move Builder Analysis

### Question 5: Percentage% of Entry Level Luxury villa

```python
entry_level=len(
    villa_dataset[villa_dataset["Price_Crore"] < 4]
)
total=len(villa_dataset)
percentage=(entry_level / total) * 100
print(f"Entry -Level Luxury Listings: {percentage:.2f}%")

### VISUALIZATION
setup_chart=("Luxury Market Composition")

plt.pie([entry_level,total - entry_level],
        labels=["Under ₹4Cr","Above ₹4Cr"],
        autopct="%1.1f%%")
plt.show()
```

### Insights:
      1.The luxury villa market is increasingly dominated by ultra-premium buyers, with 64% of properties priced above ₹4 Crore.
      2.The 36% affordable luxury segment highlights strong demand from first-time luxury buyers and investment-focused customers.

### Question 6: Area Premium Inside City

```python
area_premium=(villa_dataset.groupby(["City","Area"])["Price_Per_Sqft"].mean().reset_index())

top_areas=(area_premium.sort_values(by="Price_Per_Sqft",ascending=False).head(15))

### VISUALIZATION
setup_chart=("Top Premium Luxury Areas")

sns.barplot(data=top_areas,
            x="Price_Per_Sqft",
            y="Area",
            hue="City")
    
plt.xlabel("Price Per Sqft")    
plt.ylabel("Area")

plt.tight_layout()
plt.show()
           
```

### Insights:
       1.Certain micro-locations outperform entire cities.
       2.Premium locality branding strongly impacts sqft pricing

### Question 7: Median Buyer Profile

```python
buyer_profile = (
    villa_dataset.groupby("City") .agg({
        "Price_Crore":"median",
        "Area_Sqft":"median",
        "Bedrooms":"median"
    })
)

print("\nMedian Buyer Profile\n")

print(buyer_profile)

buyer_profile.reset_index(inplace=True)

### VISUALIZATION
setup_chart=("Median Buyer Budget by City")

sns.barplot(data=buyer_profile,
           x="City",
           y="Price_Crore")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Insights:
        1.Mumbai buyers spend significantly higher budgets.
        2.Most luxury buyers prefer 5-bedroom villas.
        3.Delhi NCR buyers prefer larger spaces

### Question 8: Duplicate Listings Analysis


```python
duplicates = villa_dataset.duplicated(subset=["City","Area","Area_Sqft","Price_Crore"],
    keep=False)

duplicate_df = villa_dataset[duplicates]

duplicate_percent = (len(duplicate_df) / len(villa_dataset)) * 100

print(f"\nDuplicate Listings: {duplicate_percent:.2f}%")

### VISUALIZATION
setup_chart=("Duplicate Listing Distortion")

duplicate_counts = pd.Series({
    "Duplicate": len(duplicate_df),
    "Unique": len(villa_dataset)-len(duplicate_df)
})

duplicate_counts.plot(
    kind="bar",
    edgecolor="white"
)

plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

### Insights:
      1.6.12% duplicate listings indicate platform redundancy.
      2.Duplicate listings distort pricing transparency.

### Question 9: Gated Community Premium

```python
gated = ( villa_dataset[villa_dataset["Gated_Community"] == "Yes"]["Price_Crore"].mean())

non_gated = (villa_dataset[villa_dataset["Gated_Community"] == "No"] ["Price_Crore"].mean())

premium = gated - non_gated

premium_df = pd.DataFrame({
 "Category": ["Gated","Non-Gated"],
 "Average_Price": [gated, non_gated]
})


setup_chart=("Gated Community Premium")

sns.barplot(data=premium_df,
        x="Category",
        y="Average_Price")

plt.ylabel("Average Price")

plt.tight_layout()
plt.show()

print(f"Gated Community Premium: ₹{premium:,.0f}")
```

### Insights:
      1.The analysis shows that non-gated luxury villas have a slightly higher average price compared to gated communities, indicating that location   and exclusivity may influence pricing more than community type.
      2.Buyers in the luxury segment may prioritize premium locations, villa size, and brand value over gated community features alone.


### Question 10: Best Entry Point Under ₹6 Crore

```python
entry_df = villa_dataset[villa_dataset["Price_Crore"] <= 60000000]

best_entry = (entry_df.groupby(["City", "Area"])["Price_Per_Sqft"].mean().sort_values().head(10).reset_index())

setup_chart=("Best Luxury Entry Markets Under ₹6Cr")

sns.barplot(data=best_entry,
           x="Price_Per_Sqft",
           y="Area",
           hue="City")
plt.xlabel("Price Per Sqft")
plt.ylabel("Area")

plt.tight_layout()
plt.show()


```

### Insights:
         1.Emerging luxury corridors offer better value below ₹6Cr.
         2.Investors can target undervalued luxury zones.
